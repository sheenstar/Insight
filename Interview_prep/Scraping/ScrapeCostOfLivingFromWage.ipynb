```python
import re
def getDataFromLocation(location_url):

    locationR = requests.get(base_url + location_url)
    locationPage = BeautifulSoup(locationR.text,"html.parser")

    expenseTable = locationPage.find('table',{'class':'expenses_table'})
    #get header
    table_header = expenseTable.find('tr')
    table_labels = []
    for t in table_header.find_all('th'):
        table_labels.append(re.sub(r"[^0-9a-zA-Z ]","",t.text))
    
    #get data
    table_rows = expenseTable.find_all('tr')
    #medical
    med_expense = table_rows[3].find_all('td')
    med_expenses = []
    for m in med_expense:
        med_expenses.append(re.sub(r"[^0-9]","",m.text.strip()))
    #total expenses
    tot_expense = table_rows[9].find_all('td')
    tot_expenses = []
    for t in tot_expense:
        tot_expenses.append(re.sub(r"[^0-9]","",t.text.strip()))
    
    return(table_labels[1:],med_expenses[1:],tot_expenses[1:])

TEST_LOCATION = '/counties/01001'
getDataFromLocation(TEST_LOCATION)
    
```
